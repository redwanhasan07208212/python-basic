### Q5:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
heads = int(input("Enter the number of heads: "))
legs = int(input("Enter the number of legs: "))

if heads > 0 and legs > 0:
    if legs % 2 == 0 and 2 * heads <= legs <= 4 * heads:
        chickens = (4 * heads - legs) // 2
        dogs = heads - chickens
        print("Number of chickens:", chickens)
        print("Number of dogs:", dogs)
    else:
        print("Invalid number of legs for the given number of heads.")
else:
    print("Heads and legs must be positive numbers.")
````

### Q6:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

### Q7:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
a = int(input("Enter the first term: "))
s_t= int(input("Enter the second term: "))
n = int(input("Enter the position of the term you want to find (n): "))

# Calculate difference
d = s_t - a

# Calculate the nth term
nth_ = a + (n - 1) * d

print("Then nth term of the arithmetic series is:",nth_)
````

### Q9:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Write your code here
# Dimensions of the milk tank (in cm)
H = 20
L = 20
B = 20
h = 3
r = 1
pi = 3.1416

# Calculate the volume of the milk tank (in cubic cm)
volume_of_tank = H * L * B

# Calculate the volume of one glass (in cubic cm)
volume_of_glass = pi * (r**2) * h

# Calculate the number of glasses
number_of_glasses = volume_of_tank / volume_of_glass

# Output the result
print(f"The number of glasses of milk that can be obtained: {int(number_of_glasses)}") #formating it.
````
