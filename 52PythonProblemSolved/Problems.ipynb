## Problem-1

```python
T = int(input('Enter the number of inputs'))

for i in range(1,T+1):
    n = int(input('Enter the number for checking even or odd'))
    if (n%2==0):
        print('Even')
    else:
        print('Odd')
```

## Problem - 2

```python
T = int(input('Enter the number of inputs: '))
n = []  

for i in range(T):
    num = int(input(f'Enter number {i+1} for checking even or odd: '))
    n.append(num) 

    if num % 2 == 0:
        print('Even')
    else:
        print('Odd')
```
